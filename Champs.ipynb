```python
cache = {}
def get_english_name(event, action, getEvent=True):
    if (event, action, getEvent) in cache:
        return cache[(event,action, getEvent)]
    result = events[(events['Event_Msg_Type'] == event) & ((events['Action_Type'] == action))].loc[:, ['Event_Msg_Type_Description','Action_Type_Description']]

    if getEvent:
        result =  result['Event_Msg_Type_Description'].iloc[0]
    else:
        result =  result['Action_Type_Description'].iloc[0]
    cache[(event, action, getEvent)] = result
    return cache[(event,action, getEvent)]

a = get_english_name(1, 44, False)
cache
```

```python
def add_english_name(df, getEvent=True):
    """df has two columns"""

    event_id = df['Event_Msg_Type']
    action_id = df['Action_Type']
    name =  get_english_name(event_id, action_id, getEvent)

    return name
    #return plays['Event_Msg_English'].apply(lambda _: get_english_name(event_id, action_id))
```

```python
def add_all_english(df):
    two = df[['Event_Msg_Type', 'Action_Type']]
    plays['Action_English'] = df.apply(lambda x: add_english_name(x, True), axis=1)
    
    plays['Event_Msg_English'] = df.apply(lambda x: add_english_name(x, False), axis=1)
    

cols = plays.columns.tolist()
cols = ['Game_id', 'Event_Num', 'Event_Msg_Type', 'Action_Type', 'Action_English', 'Event_Msg_English', 'Period', 'WC_Time', 'PC_Time', 'Option1', 'Option2', 'Option3', 'Team_id', 'Person1', 'Person2', 'Person3', 'Team_id_type', 'Person1_type', 'Person2_type', 'Person3_type']
plays = plays[cols]
add_all_english(plays)
```

```python

def get_first_five_players(players, team_id):
    for index, row in first_quarter.iterrows():
        if row['Event_Msg_Type'] == 12:
            continue;
        
        #if a substitution happens before we e v e  n      k  n                                           
        if row['Event_Msg_Type'] == 8 and len(players) < 5:
            raise Exception('WWhyyyyyy!!!!') 
        p = row['Person1']
        team = get_player_team(p)
        
        if team != None and team == team_id and p not in players:
            players.add(p)
        if len(players) >= 5:
            break
```

```python
def get_points_from_row(row):
    action = row['Event_Msg_Type']
    team = get_player_team(row['Person1'])
    if action == 1 or action == 3:
        return team, row['Option1']
    return None, 0
get_points_from_row(plays.iloc[0])
```

```python
def create_chunk(team_a, team_b, players, points_scored):
    """a chunk is a portion of a game where in between substitutions. 
    basically a stretch where the same 10 players are on the court. the idea is that 
    we can break the game up into chunks, and then its easy to tell which players 
    are on the court when points are scored to easily calculate off/def rating
    """
    chunk = {}
    chunk[team_a] = {
        'players': players[team_a],
        'points': points_scored[team_a]
    }
    chunk[team_b] = {
        'players': players[team_a],
        'points': points_scored[team_b]
    }
    return chunk
```
