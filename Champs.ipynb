```python
cache = {}
def get_english_name(event, action, getEvent=True):
    if (event, action, getEvent) in cache:
        return cache[(event,action, getEvent)]
    result = events[(events['Event_Msg_Type'] == event) & ((events['Action_Type'] == action))].loc[:, ['Event_Msg_Type_Description','Action_Type_Description']]

    if getEvent:
        result =  result['Event_Msg_Type_Description'].iloc[0]
    else:
        result =  result['Action_Type_Description'].iloc[0]
    cache[(event, action, getEvent)] = result
    return cache[(event,action, getEvent)]

a = get_english_name(1, 44, False)
cache
```

```python
def get_first_five_players(game, players, team_id):
    substituted_in = []
    for index, row in game.iterrows():
        if row['Event_Msg_Type'] == 12:
            continue;
                                         
        if row['Event_Msg_Type'] == 8:
            person_leaving = row['Person1']
            person_leaving_team = get_player_team(person_leaving)
            if person_leaving_team == team_id and person_leaving not in players:
                print('hmmmm')
                players.add(player_leaving)
            substituted_in.append(row['Person2'])
            print('WWhyyyyyy!!!!', index)
        else:
            for i in range(1, 4):
                p = row['Person'+str(i)]
                team = get_player_team(p)

                if team != None and team == team_id and p not in players and p not in substituted_in:
                    players.add(p)
                if len(players) >= 5:
                    print('done')
                    return
```

```python
def create_chunk(team_a, team_b, players, points_scored, possessions):
    """a chunk is a portion of a game where in between substitutions. 
    basically a stretch where the same 10 players are on the court. the idea is that 
    we can break the game up into chunks, and then its easy to tell which players 
    are on the court when points are scored to easily calculate off/def rating
    """
#     chunk = {'possessions': possessions}
#     chunk[team_a] = {
#         'players': players[team_a],
#         'points': points_scored[team_a]
#     }
#     chunk[team_b] = {
#         'players': players[team_b],
#         'points': points_scored[team_b]
#     }
    
    chunk = {}
    for p in players[team_a]:
        chunk[p] = {'points_for': points_scored[team_a],
                    'points_against': points_scored[team_b],
                          'possessions': possessions}
    for p in players[team_b]:
        chunk[p] = {'points_for': points_scored[team_b],
                    'points_against': points_scored[team_a],
                          'possessions': possessions}
    return chunk
```

```python
def check_possessions(game, index):
    play = game.iloc[index]
    this_team = get_player_team(play['Person1'])
    
    if index + 1 < game.shape[0]:
        next_play = game.iloc[index + 1]
        next_team = get_player_team(next_play['Person1'])
    else:
        next_play = {'Team_id': None}
        next_team = None
    if((play['Event_Msg_Type']==1) #made_shot
       or ((play['Event_Msg_Type'] == 3 #free throw
            and ((play['Action_Type'] == 10) or (play['Action_Type'] == 12) or play['Action_Type'] == 15)) #final free throw
            and (next_team != this_team))
        #missed shot, defensive rebound
       or ((play['Event_Msg_Type']==2) and (next_team != this_team))
       or (play['Event_Msg_Type']==5) #turnover
       or (play['Event_Msg_Type']==13)): #end period
        #print('POSESSSION', index)
        return 1
    return 0
```

```python
def substitute_players(player_leaving, player_entering, players):
    print('substitution')
    team = get_player_team(player_leaving)

    print('team:', team)
    print('roster', players)
    print('leaving:', player_leaving)
    print('entering:', player_entering)
    print('&&&&&&&&&&&&&&')
    players[team].remove(player_leaving)
    players[team].add(player_entering)
                       
```

```python
def get_teams_from_game(game_id):
    teams = list(plays[plays['Game_id'] == game_id]['Team_id'].value_counts().sort_values()[-2:].keys())
    if len(teams) > 2:
        print(game_id)
        raise Exception('Pls no why more than 2 teams')
    if len(teams) < 2:
        raise Exception('This should ever happen, less than 2 teams')
    return teams[0], teams[1]
#[get_teams_from_game(g) for g in game_ids]
```
